Проверим на сколько ночей (`stays_total_nights`) в среднем бронируют отели типа `City Hotel`, `Resort Hotel`.

```python
round(
bookings[['hotel', 'stays_total_nights']]\
    .groupby(by='hotel', as_index=False)\
    .mean('stays_total_nights')\
    .rename(columns={'stays_total_nights': 'mean_stays_total_nights'}), 2)
```

```python
bookings[['adults', 'children', 'babies']].describe()
```

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
round(bookings[['hotel', 'total_kids']]\
    .groupby(by='hotel', as_index=False)\
    .mean(), 2)
```

Теперь посчитаем Churn rate (отток) - процент клиентов, которые отменили бронирование. Посмотрим, как метрика связана с наличием детей у клиентов. 

```python
def has_kids(bookings):
    if bookings['total_kids'] >= 1:
        return True
    else:
        return False
```

```python
bookings['has_kids'] = bookings.apply(has_kids, axis=1)
```

```python
bookings.shape
```

```python
bookings_has = bookings[bookings.has_kids == True]
bookings_hasnt = bookings[bookings.has_kids == False]
```

```python
bookings_has[bookings_has.is_canceled == 1].shape[0] / bookings_has.shape[0]
```

```python
bookings_hasnt[bookings_hasnt.is_canceled == 1].shape[0] / bookings_hasnt.shape[0]
```

Процент оттока выше у клиентов, у которых нет детей.
